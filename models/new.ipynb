```python
# 다시 처음부터 써봄 2025-09-02
import re
import json

# Regex patterns
patterns = {
    # Website form fields
    "form_name": re.compile(r"Name:\s*(.+)", re.IGNORECASE),
    "form_company": re.compile(r"Company:\s*(.+)", re.IGNORECASE),
    "form_phone": re.compile(r"Phone:\s*([\+\d\s().-]+)", re.IGNORECASE),
    "form_email": re.compile(r"E-?Mail:\s*(.+)", re.IGNORECASE),
    "form_country": re.compile(r"Country:\s*(.+)", re.IGNORECASE),

    # General info from signatures (after greetings)
    "signature_start": re.compile(
        r"(?:Mit freundlichen Grüßen|Freundliche Grüße|Beste Grüße|Viele Grüße|Herzliche Grüße|"
        r"Liebe Grüße|Schöne Grüße|Grüße|Best regards|Kind regards|Regards|Sincerely|"
        r"Yours sincerely|Yours faithfully|Thank you|Thanks)",
        re.IGNORECASE
    ),

    # Common signature fields
    "name": re.compile(r"^\s*([A-ZÄÖÜ][a-zäöüß]+(?:\s+[A-ZÄÖÜ][a-zäöüß]+)+)", re.MULTILINE),
    "role": re.compile(r"(?:CEO|COO|CTO|CFO|Chief|Manager|Director|Head|Planning|Lead|Engineer|Sales|Marketing|Board|Technik\w*)", re.IGNORECASE),
    "company": re.compile(r"([A-Z][\w&\-,\. ]*(?:GmbH|Ltd|Inc|AG))", re.IGNORECASE),
    "company_phone": re.compile(r"(\+\d{1,3}\s?\d[\d\s-]{6,})", re.IGNORECASE), 
    "website": re.compile(r"(https?://[^\s]+|www\.[^\s]+)", re.IGNORECASE),
    "email": re.compile(r"[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,}", re.IGNORECASE),
    "personal_phone": re.compile(r"(?:Mobil|Mobile|Handy|Cell):?\s*([\+\d\s().-]+)", re.IGNORECASE),
    "company_address": re.compile(r"([A-Z][a-zA-Z\s\-]+,\s*[A-Z][a-zA-Z\s]+)", re.IGNORECASE), 
}


def clean_rubitherm_blocks(text: str, anchors=None, context=4):
    """
    Remove +/- context lines around given anchors to ignore Rubitherm's own block.
    Also removes empty lines for smoother processing.
    
    anchors: list of strings to detect and remove. e.g., ["@rubitherm.com", "rubitherm"]
    """
    if anchors is None:
        anchors = ["@rubitherm.com", "Rubitherm", "rubitherm"]  # default anchor

    lines = [line for line in text.splitlines() if line.strip()]  # 빈 줄 제거
    keep_lines = []

    for i, line in enumerate(lines):
        line_lower = line.lower()
        skip = False
        for anchor in anchors:
            anchor_lower = anchor.lower()
            # 현재 줄에 anchor 포함되거나, +/- context 줄에 포함되면 skip
            if anchor_lower in line_lower or any(anchor_lower in lines[j].lower() 
                                                for j in range(max(0, i-context), min(len(lines), i+context+1))):
                skip = True
                break
        if not skip:
            keep_lines.append(line)

    return "\n".join(keep_lines)



def extract_customer_info(email_text: str):
    results = {}
    extracted_by = None

# --------------------
# 1) Website form inquiry
# --------------------
    # Entire block extraction (Country ~ E-Mail)
    form_block_match = re.search(r"Country:.*?E-?Mail:.*", email_text, re.DOTALL | re.IGNORECASE)
    if form_block_match:
        form_block = form_block_match.group(0)

        name = patterns["form_name"].search(form_block)
        if name:
            full_name = name.group(1).strip()
            parts = full_name.split()
            if len(parts) >= 2:
                results["first_name"] = parts[0]
                results["last_name"] = " ".join(parts[1:])
            else:
                results["first_name"] = full_name

        company = patterns["form_company"].search(form_block)
        if company:
            results["company"] = company.group(1).strip()

        phone = patterns["form_phone"].search(form_block)
        if phone:
            results["customer_phone"] = phone.group(1).strip()

        email = patterns["form_email"].search(form_block)
        if email:
            results["email"] = email.group(1).strip()

        country = patterns["form_country"].search(form_block)
        if country:
            results["country"] = country.group(1).strip()

        # Message: everything after E-Mail line
        email_match = re.search(r"E-?Mail:\s*([^\s]+)", form_block, re.IGNORECASE)
        if email_match:
            end_pos = email_match.end()
            results["message"] = email_text[end_pos:].strip()

        extracted_by = "form_inquiry_extractor"
        return {
            "extracted_data": results,
            "extracted_by" : extracted_by
            }

    # --------------------
    # 2) Custumer direct inquiry
    # --------------------

    text = clean_rubitherm_blocks(email_text)

    # 모든 Rubitherm 관련 라인 제거
    lines = [line for line in text.splitlines() if "@rubitherm.com" not in line.lower() and "rubitherm" not in line.lower()]
    text_clean = "\n".join(lines)

    # signature 찾기
    signature_match = patterns["signature_start"].search(text_clean)
    if signature_match:
        sig_text = text_clean[signature_match.end():].strip()
    else:
        sig_text = text_clean

    # Name
    name = patterns["name"].search(sig_text)
    if name:
        full_name = name.group(1).strip()
        parts = full_name.split()
        if len(parts) >= 2:
            results["first_name"] = parts[0]
            results["last_name"] = " ".join(parts[1:])
        else:
            results["first_name"] = full_name

    # Role
    role = patterns["role"].search(sig_text)
    if role:
        results["role"] = role.group(0).strip()

    # Company
    company = patterns["company"].search(sig_text)
    if company:
        results["company"] = company.group(1).strip()

    # Phone
    company_phone = patterns["company_phone"].search(sig_text)
    if company_phone:
        results["company_phone"] = company_phone.group(1).strip()
    personal_phone = patterns["personal_phone"].search(sig_text)
    if personal_phone:
        results["customer_phone"] = personal_phone.group(1).strip()

    # Website
    website = patterns["website"].search(sig_text)
    if website:
        results["company_website"] = website.group(1).strip()

    # Address 
    address_lines = []
    for line in sig_text.splitlines():
        # 숫자가 1개 이상 포함된 라인, 길이가 짧지 않은 경우
        if re.search(r'\d', line) and len(line.strip()) > 6:
            address_lines.append(line.strip())

    # 여러 라인이 있을 경우, 연결해서 하나의 주소로
    if address_lines:
        results["company_address"] = ", ".join(address_lines)

    # Email: @rubitherm.com 제외
    email = patterns["email"].findall(sig_text)
    if email and "@rubitherm.com" not in email.group(0).lower():
        results["email"] = list(set(email))

    extracted_by = "direct_email_extractor"

    return {"extracted_data": results, "extracted_by": extracted_by}

```
