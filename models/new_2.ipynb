```python
import re
import json

# --------------------
# Regex patterns
# --------------------
patterns = {
    "form_name": re.compile(r"^\s*Name:\s*(.+)", re.IGNORECASE | re.MULTILINE),
    "form_company": re.compile(r"^\s*Company:\s*(.+)", re.IGNORECASE | re.MULTILINE),
    "form_phone": re.compile(r"^\s*Phone:\s*([\+\d\s().-]+)", re.IGNORECASE | re.MULTILINE),
    "form_email": re.compile(r"^\s*E-?Mail:\s*([^\s]+)", re.IGNORECASE | re.MULTILINE),
    "form_country": re.compile(r"^\s*Country:\s*(.+)", re.IGNORECASE | re.MULTILINE),

    # Greetings (signature 시작 지점)
    "signature_start": re.compile(
        r"(?:Mit freundlichen Grüßen|Freundliche Grüße|Beste Grüße|Viele Grüße|Herzliche Grüße|"
        r"Liebe Grüße|Schöne Grüße|Grüße|Grüsse|Greusse|Best regards|Kind regards|Regards|Sincerely|"
        r"Yours sincerely|Yours faithfully|Thank you|Thanks)",
        re.IGNORECASE
    ),

    # Company candidates 
    "company": re.compile(r"\b([A-Z][A-Za-z&\-,\. ]{1,29}\b(?:GmbH|Ltd|Inc|AG)?)", re.IGNORECASE),

    # URL
    "website": re.compile(r"(https?://[^\s]+|www\.[^\s]+)", re.IGNORECASE),

    # Email (multiple)
    "email": re.compile(r"[A-Za-z0-9._%+\-]+@[A-Za-z0-9.\-]+\.[A-Za-z]{2,}", re.IGNORECASE),
}

# Signature patterns
NAME_LINE = re.compile(
    r"(?im)^\s*(?:Dr\.|Prof\.|Dipl\.-?Ing\.|Ing\.|Mr\.|Ms\.|Mrs\.|Herr|Frau|Eng\.)?\s*"
    r"([A-ZÄÖÜ][\w’'\-äöüßÄÖÜ]+)\s+([A-ZÄÖÜ][\w’'\-äöüßÄÖÜ]+)\s*$"
)
ROLE_LINE = re.compile(
    r"(?im)^\s*(?:(?!.*@).)*(?:\b("
    r"CEO|COO|CTO|CFO|Chief|Manager|Director|Head|Lead|Engineer|Engineering|Sales|Marketing|"
    r"Research(?:er)?|Supervisor|Chairman|Board|President|Professor|Consultant|"
    r"Technik\w*|Leiter|Leitung|Geschäftsführer|Vorstand"
    r")\b).*?$"
)
# Tel/Mobile Label phone numbers
LABELED_PHONE = re.compile(
    r"(?im)^\s*(?:Tel(?:efon)?|Phone|Telefon|Mobile|Mobil|Handy|Cell)\s*[:\-]?\s*([+()\d][\d\s().\-]{6,})\s*$"
)
# No label phone number
BARE_INTL_PHONE = re.compile(
    r"(?im)^(?!.*Fax)(?!.*FAX)\s*\+?\d{1,3}[\s\-]?\d[\d\s\-\(\)]{6,}\s*$"
)

# Address candidates: "City, Country" 
CITY_COUNTRY = re.compile(r"(?im)^\s*[A-Z][\w\s\-.]+,\s*[A-Z][\w\s\-.]+\s*$")
# Address candidates: "StreetName 12" or "D-12345 Berlin" Combination of Numbers and characters (tel/fax/web/email/company names excluded)
NUMBERED_ADDRESS_LINE = re.compile(r"(?im)^\s*(?=.*\d)(?!.*@)(?!.*Fax)(?!.*TEL)(?!.*Tel)(?!.*Phone)(?!.*www)(?!.*http)(?!.*GmbH|.*AG|.*Ltd|.*Inc)\S.+$")

# --------------------
# Utils
# --------------------
def uniq(seq, key=lambda x: x.lower() if isinstance(x, str) else x):
    seen = set()
    out = []
    for x in seq:
        k = key(x)
        if k in seen:
            continue
        seen.add(k)
        out.append(x.strip() if isinstance(x, str) else x)
    return out

def strip_empty_lines(text: str) -> str:
    return "\n".join([ln for ln in text.splitlines() if ln.strip()])

def clean_rubitherm_blocks(text: str, anchors=None, context=4):
    """
    - no empty lines
    - delete lines around anchors ±context
    """
    if anchors is None:
        anchors = ["@rubitherm.com", " Rubitherm ", "<rubitherm.com>", "(rubitherm.com)", "rubitherm"]

    # Eliminate empty lines
    lines = [ln for ln in text.splitlines() if ln.strip()]
    keep = []
    n = len(lines)

    def has_anchor(idx):
        low = max(0, idx - context)
        high = min(n, idx + context + 1)
        block = " ".join(lines[low:high]).lower()
        return any(a.lower() in block for a in anchors)

    for i, line in enumerate(lines):
        if has_anchor(i):
            continue
        keep.append(line)

    return "\n".join(keep)

# --------------------
# Signature Parser (list)
# --------------------
def parse_signature_block(sig_text: str):
    # Filtering Rubitherm once more
    lines = []
    for ln in sig_text.splitlines():
        low = ln.lower()
        if "rubitherm" in low or "@rubitherm.com" in low:
            continue
        lines.append(ln)
    sig = "\n".join(lines).strip()

    out = {
        "first_name": None,
        "last_name": None,
        "emails": [],
        "phones": [],
        "companies": [],
        "roles": [],
        "company_addresses": [],
        "websites": [],
    }

    # emails (No Rubitherm)
    emails = [e for e in patterns["email"].findall(sig) if "@rubitherm.com" not in e.lower()]
    out["emails"] = uniq(emails)

    # phones, no Fax
    phones = []
    phones += LABELED_PHONE.findall(sig)
    phones += [m.group(0).strip() for m in BARE_INTL_PHONE.finditer(sig)]
    # Quick regularization (no empty space)
    phones = [re.sub(r"\s{2,}", " ", p) for p in phones]
    out["phones"] = uniq(phones)

    # companies
    comps = patterns["company"].findall(sig)
    out["companies"] = uniq(comps)

    # websites
    websites = [w.group(1) if isinstance(w, re.Match) else w for w in patterns["website"].findall(sig)]
    # re.findall returns list of strings (because of group), normalize:
    if websites and isinstance(websites[0], tuple):
        websites = [w[0] for w in websites]
    out["websites"] = uniq(websites)

    # roles (entire lines are preserved)
    out["roles"] = uniq([m.group(0).strip() for m in ROLE_LINE.finditer(sig)])

    # names (Not in company, role, email)
    for ln in sig.splitlines():
        if not ln.strip():
            continue
        if patterns["email"].search(ln) or patterns["company"].search(ln) or ROLE_LINE.search(ln):
            continue
        nm = NAME_LINE.match(ln)
        if nm:
            out["first_name"], out["last_name"] = nm.group(1), nm.group(2)
            break

    # addresses
    addr_lines = []
    addr_lines += [m.group(0).strip() for m in CITY_COUNTRY.finditer(sig)]
    addr_lines += [m.group(0).strip() for m in NUMBERED_ADDRESS_LINE.finditer(sig)]
    # (Not in company, role, email)
    pruned = []
    for a in addr_lines:
        low = a.lower()
        if patterns["email"].search(a) or "fax" in low or "tel" in low or "phone" in low or "www" in low or "http" in low:
            continue
        pruned.append(a)
    out["company_addresses"] = uniq(pruned)

    return out

# --------------------
# Main Extraction Function
# --------------------
def extract_customer_info(email_text: str):
    results = {}
    extracted_by = None

    # 0) no empty lines
    email_text = strip_empty_lines(email_text)

    # --------------------
    # 1) Website form inquiry
    # --------------------
    form_block_match = re.search(r"(?si)Country:\s*.+?E-?Mail:\s*[^\s]+", email_text, re.DOTALL | re.IGNORECASE)
    if form_block_match:
        form_block = form_block_match.group(0)

        # single value
        name = patterns["form_name"].search(form_block)
        if name:
            full = name.group(1).strip()
            parts = full.split()
            if len(parts) >= 2:
                results["first_name"] = parts[0]
                results["last_name"] = " ".join(parts[1:])
            else:
                results["first_name"] = full

        company = patterns["form_company"].search(form_block)
        if company:
            results["company"] = company.group(1).strip()

        phone = patterns["form_phone"].search(form_block)
        if phone:
            results["customer_phone"] = phone.group(1).strip()

        email_m = patterns["form_email"].search(form_block)
        if email_m:
            results["email"] = email_m.group(1).strip()

        country = patterns["form_country"].search(form_block)
        if country:
            results["country"] = country.group(1).strip()

        # result in the List 
        results["emails"] = [results["email"]] if "email" in results else []
        results["phones"] = [results["customer_phone"]] if "customer_phone" in results else []
        results["companies"] = [results["company"]] if "company" in results else []
        results["roles"] = []
        results["company_addresses"] = []
        results["websites"] = []

        # Message: after form block to the end 
        results["message"] = email_text[form_block_match.end():].strip()

        extracted_by = "form_inquiry_extractor"
        return {"extracted_data": results, "extracted_by": extracted_by}

    # --------------------
    # 2) Direct customer email
    # --------------------
    # Delete Rubitherm block 
    text = clean_rubitherm_blocks(email_text, anchors=["@rubitherm.com", " rubitherm ", "<rubitherm.com>", "(rubitherm.com)", "rubitherm"], context=6)

    # Extract Signature block
    m = patterns["signature_start"].search(text)
    sig_text = text[m.end():].strip() if m else text

    sig_parsed = parse_signature_block(sig_text)

    # Single value field 
    if sig_parsed["first_name"]:
        results["first_name"] = sig_parsed["first_name"]
    if sig_parsed["last_name"]:
        results["last_name"] = sig_parsed["last_name"]

    # List value field
    results["emails"] = sig_parsed["emails"]
    results["phones"] = sig_parsed["phones"]
    results["companies"] = sig_parsed["companies"]
    results["roles"] = sig_parsed["roles"]
    results["company_addresses"] = sig_parsed["company_addresses"]
    results["websites"] = sig_parsed["websites"]

    # 호환용 단일 값(있으면 첫 값)
    """    if results.get("companies"):
        results["company"] = results["companies"][0]
    if results.get("phones"):
        results["customer_phone"] = results["phones"][0]
    if results.get("emails"):
        results["email"] = results["emails"][0]"""

    extracted_by = "direct_email_extractor"
    return {"extracted_data": results, "extracted_by": extracted_by}
```
